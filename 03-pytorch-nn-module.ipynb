```python
class Model(nn.Module):

    def __init__(self,num_features):
        super().__init__()

        self.neural_network = nn.Sequential(
            nn.Linear(num_features,3),
            nn.ReLU(),
            nn.Linear(3,1),
            nn.Sigmoid()
        )

    def forward(self, features):
        return self.neural_network(features)
```

```python
features = torch.rand(10,5)
```

```python
features
```

```python
model = Model(features.shape[1])
```

```python
features.shape[1]
```

```python
model(features)
```

```python
# To get the weights of the FIRST linear layer (num_features -> 3)
print(model.neural_network[0].weight)

# To get the weights of the SECOND linear layer (3 -> 1)
print(model.neural_network[2].weight)
```

```python
# To get the bias of the FIRST linear layer (num_features -> 3)
print(model.neural_network[0].bias)

# To get the bias of the SECOND linear layer (3 -> 1)
print(model.neural_network[2].bias)
```

```python
summary(model, input_size=(10,5))
```
