## Differentiation of y = $x^2$, z = sin(y). Find dz/dx

```python
import math
```

```python
x = 3
```

```python
dz_dy = math.cos(x**2)
dy_dx = 2 * x
dz_dx = dz_dy * dy_dx
```

```python
dz_dy,dy_dx
```

```python
dz_dx
```

## Differentiation of y = $x^2$, z=sin(y), u = $e^z$. Find du/dx

```python
x = 3
```

```python
z = math.sin(x**2)
dy_dx = 2 * x
dz_dy = math.cos(x**2)
du_dz = math.exp(z)
du_dx = dy_dx * dz_dy * du_dz
```

```python
du_dx
```

```python
z
```
